```python
import pandas as pd
```

```python
df = pd.read_csv('../data/vine_table.csv')
df.head()
```

```python
df = df.loc[df['total_votes'] >= 20]
df.head()
```

```python
df['helpful_over_total_votes'] = df['helpful_votes']/df['total_votes']
df.head()
```

```python
df = df.loc[df['helpful_over_total_votes'] >= 0.5]
df.head()
```

```python
vine_Y_df = df.loc[df['vine'] == 'Y']
vine_Y_df.head()
```

```python
vine_N_df = df.loc[df['vine'] == 'N']
vine_N_df.head()
```

```python
total_Y = len(vine_Y_df.review_id)
total_N = len(vine_N_df.review_id)
total_Y_5S = len(vine_Y_df.loc[vine_Y_df['star_rating'] == 5].review_id)
total_N_5S = len(vine_N_df.loc[vine_N_df['star_rating'] == 5].review_id)
pct_Y_5S = total_Y_5S / total_Y
pct_N_5S = total_N_5S / total_N
print("Vine reviews: " + str(total_Y))
print("Other reviews: " + str(total_N))
print("Vine 5-star reviews: " + str(total_Y_5S))
print("Other 5-star reviews: " + str(total_N_5S))
print("Vine 5-star percentage of reviews: " + str(pct_Y_5S))
print("Other 5-star percentage of reviews: " + str(pct_N_5S))
```
